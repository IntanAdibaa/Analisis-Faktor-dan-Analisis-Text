Uji kesesuaian data menggunakan kelas calculate_kmo pada factor_analyzer library

```python
kmo_all,kmo_model=calculate_kmo(dataset)
kmo_model
```

Terapkan analisis faktor ke kumpulan data menggunakan kelas FactorAnalyzer di factor_analyzer library

```python
fa = FactorAnalyzer(n_factors = 3, rotation="varimax")
fa.fit(dataset)
```

Cek loadings

```python
loadings_output = pd.DataFrame(fa.loadings_,
                               index=dataset.columns)
loadings_output
```

Menentukan Jumlah Faktor

```python
fa = FactorAnalyzer(rotation = 'varimax',n_factors=dataset.shape[1])
fa.fit(dataset)
ev,_=fa.get_eigenvalues()

factor_values=np.arange(dataset.shape[1])+1

plt.figure(figsize=(15,8))

plt.plot(factor_values,ev,'o-',linewidth=2)
plt.title("Scree Plot")
plt.xlabel("Factors")
plt.ylabel("Eigen Value")
plt.grid()
```

Berdasarkan scree plot diatas, n factornya adalah 4

Maka dilakukan lagi penerapan analisis faktor ke kumpulan data menggunakan kelas FactorAnalyzer di factor_analyzer library dengan N_factor = 4

```python
fa = FactorAnalyzer(n_factors = 4, rotation="varimax")
fa.fit(dataset)
```
